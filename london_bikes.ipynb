```python
# mapping values of season and weather values
# season map
season_dict = {
    '0.0': 'Spring',
    '1.0': 'Summer',
    '2.0': 'Autumn',
    '3.0': 'Winter'
}
# weather map
weather_dict = {
    '1.0': 'Clear',
    '2.0': 'Scattered clouds',
    '3.0': 'Broken clouds',
    '4.0': 'Cloudy',
    '7.0': 'Rainy',
    '10.0': 'Storm',
    '26.0': 'Snowfall'
}

bikes_df.season = bikes_df.season.astype('str')
bikes_df.season = bikes_df.season.map(season_dict)

bikes_df.weather_code = bikes_df.weather_code.astype('str')
bikes_df.weather_code = bikes_df.weather_code.map(weather_dict)
```
